```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

fight_stats = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\raw\ufc_fight_details.csv")
fight_stats.head(5)
```

```python
fight_stats.columns = fight_stats.columns.str.replace("totals_perround_", "")
fight_stats.columns = fight_stats.columns.str.replace("round_", "r")  
fight_stats.columns = fight_stats.columns.str.replace("significant_strikes_perr", "sg_strk_")  
fight_stats.columns = fight_stats.columns.str.replace("is_title_bout", "title") 
fight_stats.columns = fight_stats.columns.str.replace("time_format", "set_rounds")
fight_stats.columns = fight_stats.columns.str.replace("time", "last_rnd_time")


fight_stats = fight_stats.drop(columns=[
    "event_url",
    "perf_bonus",
    "details", "referee",
    "fight_url",
    "nickname_f1","nickname_f2",
    "fighter_f1_url","fighter_f2_url",
     "r1_fighter_f1","r1_fighter_f2", "r1_fighter_f1_url","r1_fighter_f2_url",
    "r2_fighter_f1","r2_fighter_f2", "r2_fighter_f1_url","r2_fighter_f2_url",
    "r3_fighter_f1","r3_fighter_f2", "r3_fighter_f1_url","r3_fighter_f2_url",
    "r4_fighter_f1","r4_fighter_f2", "r4_fighter_f1_url","r4_fighter_f2_url",
    "r5_fighter_f1","r5_fighter_f2", "r5_fighter_f1_url","r5_fighter_f2_url",
    "sg_strk_r1_fighter_f1", "sg_strk_r1_fighter_f2", "sg_strk_r1_fighter_f1_url", "sg_strk_r1_fighter_f2_url",
    "sg_strk_r2_fighter_f1", "sg_strk_r2_fighter_f2", "sg_strk_r2_fighter_f1_url", "sg_strk_r2_fighter_f2_url",
    "sg_strk_r3_fighter_f1", "sg_strk_r3_fighter_f2", "sg_strk_r3_fighter_f1_url", "sg_strk_r3_fighter_f2_url",
    "sg_strk_r4_fighter_f1", "sg_strk_r4_fighter_f2", "sg_strk_r4_fighter_f1_url", "sg_strk_r4_fighter_f2_url",
    "sg_strk_r5_fighter_f1", "sg_strk_r5_fighter_f2", "sg_strk_r5_fighter_f1_url", "sg_strk_r5_fighter_f2_url"
                             ])


fight_stats["outcome_f1"] = fight_stats["outcome_f1"].map({"W": 1, "L": 0, "NC": 2, "D":3}).astype("Int64")
fight_stats["outcome_f2"] = fight_stats["outcome_f2"].map({"W": 1, "L": 0, "NC": 2, "D":3}).astype("Int64")
fight_stats["set_rounds"] = fight_stats["set_rounds"].astype(str).str.extract(r"(\d+)")
fight_stats['last_rnd_time'] = pd.to_timedelta( fight_stats['last_rnd_time'].apply(lambda x: f"00:{x}" if len(x.split(':')) == 2 else x)).dt.total_seconds().astype(int)

fight_stats["round"] = fight_stats["round"].astype(int)

percent_columns = [
    col for col in fight_stats.select_dtypes(include=['object', 'category']).columns
    if fight_stats[col].astype(str).str.contains('%').any() 
    and col.lower() != "event_name"  # Avoid event_name
]
for col in percent_columns:
    cleaned_series = fight_stats[col].astype(str).str.rstrip('%').str.strip()
    fight_stats[col] = pd.to_numeric(cleaned_series, errors='coerce') / 100  

# 2. "of" Columns Conversion (e.g., strikes landed '15 of 30')
of_columns = [
    col for col in fight_stats.select_dtypes(include=['object', 'category']).columns
    if fight_stats[col].astype(str).str.contains(' of ').any() 
    and col.lower() != "event_name"  # Avoid event_name
]
for col in of_columns:
    split_df = fight_stats[col].astype(str).str.split(' of ', expand=True)
    numerator = pd.to_numeric(split_df[0], errors='coerce')
    denominator = pd.to_numeric(split_df[1], errors='coerce')
    fight_stats[col] = numerator / denominator.replace(0, np.nan)

# 3. Time Columns Conversion (e.g., '5:00')
time_columns = [
    col for col in fight_stats.select_dtypes(include=['object', 'category']).columns
    # Uses regex to ONLY match structural timestamps like 5:00, ignoring fight names with colons
    if fight_stats[col].astype(str).str.contains(r'^\d+:\d+$').any() 
    and col.lower() != "event_name"  # Avoid event_name
]
for col in time_columns:
    time_parts = fight_stats[col].astype(str).str.split(':')
    fight_stats[col] = time_parts.apply(
        lambda x: int(x[0]) * 60 + int(x[1]) if isinstance(x, list) and len(x) == 2 and x[0].isdigit() and x[1].isdigit() else None
    )




fight_stats["tot_fight_time"] = fight_stats["last_rnd_time"] + (fight_stats["round"] - 1) * 300

fight_stats.insert(9, "tot_fight_time", fight_stats.pop("tot_fight_time"))


print(len(fight_stats.columns))
print((fight_stats.columns.dtype))
print((fight_stats.columns))
#fight_stats_null = fight_stats[fight_stats["outcome_f1"].isnull()]
fight_stats
```
