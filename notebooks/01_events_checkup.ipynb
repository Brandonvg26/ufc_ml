# **EVENTS CHECK & CONTEXT**

```python
import pandas as pd

events = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\raw\ufc_events.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

events["event_date"] = pd.to_datetime(events["event_date"], errors="coerce")

events["event_date"] = events["event_date"].dt.strftime("%m/%d/%Y")


print(events.columns)
events.head(3)
```

## CATEGORICAL COLUMNS CHECK

```python
#result_raw weight_class	method	method_detail	round	
print(f"Result types ({(events['result_raw'].nunique())}): {list(events['result_raw'].unique())}")
print(f"Weight types ({(events['weight_class'].nunique())}): {list(events['weight_class'].unique())}")
print(f"Method types ({(events['method'].nunique())}): {list(events['method'].unique())}")
print(f"Method Detail types ({(events['method_detail'].nunique())}): {list(events['method_detail'].unique())}")
print(f"Round types ({(events['round'].nunique())}): {list(events['round'].unique())}")
print(f"Latest Event: {(events['event_date'].max())}")
print(f"Earliest Event: {(events['event_date'].min())}")
print(f"\n          NULL VALUES\n{events.isna().sum()}")
```

### CHECK OF NULL VALUES

```python
null_detail = events[events["method_detail"].isnull()]
print(f"Method types ({(null_detail['method'].nunique())}): {list(null_detail['method'].unique())}")
null_detail.head(3)
```

```python
null_winner = events[events["winner"].isnull() ]#& (events["result_raw"]=="nc/nc")]
print(f"Method types ({(null_winner['method'].nunique())}): {list(null_winner['method'].unique())}")
print(f"Result types ({(null_winner['result_raw'].nunique())}): {list(null_winner['result_raw'].unique())}")
null_winner.head(3)
```

#### CONTEXT OF NULL VALUES

##### * NULL DETAILS - FIGHT WENT WITHOUT EXTRA REQ. DATA BESIDES ***-METHOD-***
##### * NULL WINNER  - FIGHT WENT DRAW/DRAW OR ARE NC/NC (NO CONTEST - fight occured but was stopped due to a fighter being fauled/injured or not passing tests)

# **EVENTS LIST**

```python
fights_list = events[["event_name","event_date","fight_url"]].drop_duplicates()

fights_list.to_csv(r"C:\Git projects\ufc_ml\ufc_data\lists\list_fights.csv", index=False)

fights_list
```

# **FIGHTERS LIST**

```python

fighters = events[["fighter1","fighter1_url","fighter2","fighter2_url"]]

part1 = fighters[["fighter1", "fighter1_url"]].rename(
    columns={"fighter1": "fighter", "fighter1_url": "fighter_url"}
)

part2 = fighters[["fighter2", "fighter2_url"]].rename(
    columns={"fighter2": "fighter", "fighter2_url": "fighter_url"}
)

fighters = pd.concat([part1, part2], ignore_index=True)

fighters = fighters.drop_duplicates(ignore_index=True)

pd.set_option('display.max_colwidth', None)

fighters.to_csv(r"C:\Git projects\ufc_ml\ufc_data\lists\list_fighters.csv", index=False)
(fighters)
```
