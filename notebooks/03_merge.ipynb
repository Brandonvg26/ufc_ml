```python
import pandas as pd

fighters = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\slv\data_stats_fighters.csv")
display(fighters)
events = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\lists\list_fights.csv")
display(events)
fight_stats = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\slv\fight_details_clean.csv")
display(fight_stats)
```

```python
fight_data = fight_stats.copy()
fight_data = fight_data.drop(columns=["event_date_x", "event_date_y", "event_date"], errors="ignore")

events_clean = events[["event_name", "event_date"]].drop_duplicates(subset=["event_name"], keep="first")
fight_data = fight_data.merge(events_clean, on="event_name", how="left")

fight_data = fight_data.merge(
    fighters.add_suffix("_f1"), 
    left_on="fighter_f1", 
    right_on="fighter_f1", 
    how="left"
)

fight_data = fight_data.merge(
    fighters.add_suffix("_f2"), 
    left_on="fighter_f2", 
    right_on="fighter_f2", 
    how="left"
)

fight_data["event_date"] = pd.to_datetime(fight_data["event_date"], errors="coerce")
fight_data["DOB_f1"] = pd.to_datetime(fight_data["DOB_f1"], errors="coerce")
fight_data["DOB_f2"] = pd.to_datetime(fight_data["DOB_f2"], errors="coerce")

fight_data["age_at_fight_f1"] = (fight_data["event_date"] - fight_data["DOB_f1"]).dt.days // 365
fight_data["age_at_fight_f2"] = (fight_data["event_date"] - fight_data["DOB_f2"]).dt.days // 365

fight_data = fight_data.drop(columns=["DOB_f1", "DOB_f2"], errors="ignore")

display(fight_data)
```

```python
fight_data.to_csv(r"C:\Git projects\ufc_ml\ufc_data\obt\fights_data.csv", index=False)
```
