```python
import pandas as pd

raw = pd.read_csv(r"C:\Git projects\ufc_ml\ufc_data\fighter_stats.csv", sep=",")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

print(raw.columns)
fighters = raw.copy()
fighters = fighters[["fighter","raw_stats_json"]]

fighters
```

```python
import json
import pandas as pd

# 1. Convert the JSON string column into actual Python dictionaries
fighters["raw_stats_json"] = fighters["raw_stats_json"].apply(json.loads)

# 2. Flatten the dictionaries into a new DataFrame
stats_df = pd.json_normalize(fighters["raw_stats_json"])

# 3. Combine with the original fighter names and drop the old JSON column
fighters = fighters[["fighter"]].join(stats_df)

fighters["DOB"] = pd.to_datetime(fighters["DOB"], errors="coerce")

fighters["DOB"] = fighters["DOB"].dt.strftime("%m/%d/%Y")

fighters.to_csv(r"C:\Git projects\ufc_ml\ufc_data\data_stats_fighters.csv", index=False)
```
